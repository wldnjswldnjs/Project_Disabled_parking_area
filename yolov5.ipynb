```python
from glob import glob

img_list1 = glob('/home/lab34/yolov5/final_dataset_second/images/*.jpg') # Î™®Îì† Ïù¥ÎØ∏ÏßÄÎ•º Í∏ÅÏñ¥Ïò®Îã§.
img_list2 = glob('/home/lab34/yolov5/final_dataset_second/images/*.JPG')
img_list = img_list1 + img_list2
print(len(img_list1)) # Î™®Îì† Ïù¥ÎØ∏ÏßÄÏùò Í∞úÏàò ÌôïÏù∏
print(len(img_list2))
print(len(img_list))
```

```python
!pip install sklearn
```

```python
from sklearn.model_selection import train_test_split

train_img_list, val_img_list= train_test_split(img_list, test_size=0.2, random_state=2000)

print(len(train_img_list), len(val_img_list))
```

```python
!pwd
```

```python
# train, val Ïù¥ÎØ∏ÏßÄ Í≤ΩÎ°úÎ•º txt ÌååÏùºÎ°ú Ï†ÄÏû•

with open('/home/lab34/yolov5/final_dataset_second/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/home/lab34/yolov5/final_dataset_second/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```

```python
# data.yaml ÌååÏùº ÏàòÏ†ï

import yaml

with open('/home/lab34/yolov5/final_dataset_second/data.yaml', 'r') as f:
  data = yaml.safe_load(f) # json Î∂ÄÎ•¥Îäî ÌòïÏãùÍ≥º Í∞ôÏùå

print(data)

data['train'] = '/home/lab34/yolov5/final_dataset_second/train.txt'
data['val'] = '/home/lab34/yolov5/final_dataset_second/val.txt'

with open('/home/lab34/yolov5/final_dataset_second/data.yaml', 'w') as f:
  yaml.dump(data, f)

print(data)
```
